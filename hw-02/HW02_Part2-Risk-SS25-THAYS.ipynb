### Graph Creation

&#9989; **Question 3.2 (7 points)**: Create a NetworkX graph object from the territory definitions. The territories are the nodes of the graph. Implement a function that builds the graph and adds appropriate attributes for continents. _Hint:_ Look up the documentation of `add_node`, what are the possible inputs? Don't forget to connect each territory to its neighbors. Make sure to add plenty of comments in your code.

```python
def create_risk_graph(territories):
    """
    Create a NetworkX graph from the Risk territory definitions.
    
    Parameters:
        territories (dict): Dictionary containing territory and connection information
        
    Returns:
        G (nx.Graph): NetworkX graph representing the Risk board
    """
    G = nx.Graph()

    # Defines a series of colors used for the various regions of the game, as well as an indexer value that will allow us to assign color.
    colors = ['Red', 'Yellow', 'Blue', 'Orange', 'Green', 'Purple']
    i = 0
    # Iterates through the various regions within the territory list.
    for region in territories.values():
        # Iterates through the countries in the region.
        for country in region.keys():
            G.add_node(country, color=colors[i])
            # Iterates through the adjacent values in the dictionary.
            for adj in region[country]:
                G.add_edge(country, adj)
        i += 1
    return G

# Create the graph
G = create_risk_graph(risk_territories)
```

### Graph Visualization

&#9989; **Question 3.3 (4 points)**: Create a visualization of the Risk board graph. You can use AI to help you with this. The function below should:

- make a plot of the risk graph.
- color nodes by continent
- allow the user to choose a type of layout. _Hint:_ search networkx documentation for the possible layout. Min 5 layouts should be available.
- print the name of each territory on the respective node
- make a legend for the graph 
- have very descriptive comments 
- increase the size of each node from the default value  
- increase the font size of label from the default value

```python
from matplotlib.lines import Line2D

def visualize_risk_graph(G): # Created using the help of Grok AI.
    """
    Create a visualization of the Risk board graph with continents colored differently.
    
    Parameters:
        G (nx.Graph): NetworkX graph of the Risk board
    """

    # Defines the options for the user-given values.
    options = ["spring", "circular", "spectral", "random", "kamada kawai"]

    # Displays the menu
    print("Select a layout:")
    for i, option in enumerate(options, 1):  # Start numbering at 1
        print(f"{i}. {option}")

    # Unpacks the color values we stored earlier into a list we can use to set up the graph correctly.
    node_colors = [G.nodes[node]['color'] for node in G.nodes()]
    
    # Get user input
    while True:
        try:
            choice = int(input("Enter the number of your choice (1-5): "))
            if 1 <= choice <= len(options):
                break  # Valid choice, exit loop
            else:
                print(f"Please enter a number between 1 and {len(options)}.")
        except ValueError:
            print("Please enter a valid number.")
    if choice == 1:
        pos = nx.spring_layout(G)
    elif choice == 2:
        pos = nx.circular_layout(G)
    elif choice == 3:
        pos = nx.spectral_layout(G)
    elif choice == 4:
        pos = nx.random_layout(G)
    elif choice == 5:
        pos = nx.kamada_kawai_layout(G)
    nx.draw(G, pos, node_color = node_colors, with_labels=True, node_size = 150, font_size = 10)

    # Create legend entries (one for each unique color)
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label='North America', markerfacecolor='red', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='South America', markerfacecolor='yellow', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Europe', markerfacecolor='blue', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Africa', markerfacecolor='orange', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Asia', markerfacecolor='green', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Australia', markerfacecolor='purple', markersize=10)
    ]

    # Add the legend
    plt.legend(handles=legend_elements, loc='lower left')

    # Show the plot
    plt.title("Graph with Legend")
    plt.show()
    
# Visualize the graph
visualize_risk_graph(G)
```
